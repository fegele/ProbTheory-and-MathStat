```python
import math

def count_combinations(n, k):
    return int(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))

def probability(event_outcomes, total_outcomes):
    return event_outcomes/total_outcomes
```

```python
### 1a
n_deck = 52     # общее число карт в колоде
n_cross = 13    # число карт крести в колоде
k_drawn = 4     # число вытянутых карт

# число благоприятствующих исходов
cross_comb = count_combinations(n_cross, k_drawn)
print(f'{cross_comb=}')

# общее число исходов
all_comb = count_combinations(n_deck, k_drawn)
print(f'{all_comb=}')

# вероятность
cross_prob = probability(cross_comb, all_comb)
print(f'{cross_prob=}')
```

```python
# 2-й способ: вычисление вероятности произведения зависимых событий
# вероятность вытянуть первую карту крести из колоды: 13/52
# вероятность вытянуть вторую карту крести из колоды: 12/51
# вероятность вытянуть третью карту крести из колоды: 11/50
# вероятность вытянуть червёртую карту крести из колоды: 10/49
print((13/52)*(12/51)*(11/50)*(10/49))

# ответ идентичен:
```

```python
### 2
n_button = 10     # число кнопок
n_code = 3        # цифр в коде
k_pressed = 3     # число нажатых одновременно кнопок

# число благоприятствующих исходов
right_comb = count_combinations(n_code, k_pressed)
print(f'{right_comb=}')

# количество сочетаний трёх цифр на кнопках, нажатых одновременно
button_comb = count_combinations(n_button, k_pressed)
print(f'{button_comb=}')

# вероятность
code_prob = probability(right_comb, button_comb)
print(f'{code_prob=}')
```
