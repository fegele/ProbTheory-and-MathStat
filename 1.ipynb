```python
import math

def count_combinations(n, k):
    return math.factorial(n)/(math.factorial(k)*math.factorial(n-k))

def count_permutations(n, k):
    return math.factorial(n)/math.factorial(n-k)

def probability(event_outcomes, total_outcomes):
    return event_outcomes/total_outcomes
```

```python
### 1a
n_deck = 52     # общее число карт в колоде
n_cross = 13    # число карт крести в колоде
k_drawn = 4     # число вытянутых карт
# число благоприятствующих исходов
cross_comb = count_combinations(n_cross, k_drawn)

# общее число исходов
all_comb = count_combinations(n_deck, k_drawn)

# вероятность
cross_prob = probability(cross_comb, all_comb)
print(f'{cross_prob=}')
```

```python
### 1b
n_ace = 4       # число тузов в колоде

# благоприятствующие исходы:
event_prob = 0

# 1) 1 туз и 3 любые карты
ace_comb = count_combinations(n_ace, 1)
card_comb = count_combinations(n_deck - n_ace, 3)
event_prob += ace_comb*card_comb

# 2) 2 туза и 2 любые карты
ace_comb = count_combinations(n_ace, 2)
card_comb = count_combinations(n_deck - n_ace, 2)
event_prob += ace_comb*card_comb

# 3) 3 туза и 1 любая карта
ace_comb = count_combinations(n_ace, 3)
card_comb = count_combinations(n_deck - n_ace, 1)
event_prob += ace_comb*card_comb

# 4) 4 туза
ace_comb = count_combinations(n_ace, 4)
card_comb = count_combinations(n_deck - n_ace, 0)
event_prob += ace_comb*card_comb

# вероятность
ace_prob = probability(event_prob, all_comb)
print(f'{ace_prob=}')
```
