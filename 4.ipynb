```python
a = 200
b = 800

# среднее значение:
M = (a + b) / 2
print(f'{M=}')

# дисперсия
D = (b - a) ** 2 / 12
print(f'{D=}')
```

![4-2](4-2.png)

```python
D = 0.2
a = 0.5

# находим правую границу из формулы дисперсии равномерного распределения:
b = a + np.sqrt(12 * D)
print('b=', format(b, ".3g"))

# среднее значение:
M = (a + b) / 2
print('M=', format(M, ".3g"))
```

```python
# a) больше 182 см >> больше, чем +σ
p = (100 - 68) / 2
print(format(p, ".3g"), '%')
```

```python
# в) от 166 см до 190 см >> между -σ и +2σ
p = 95.4 - (95.4 - 68) / 2
print(format(p, ".3g"), '%')
```

```python
# е) не выше 150 см и не ниже 190 см >> меньше, чем -3σ, и больше, чем +2σ
p = 100 - 95.4 - (99.72 - 95.4) / 2
print(format(p, ".3g"), '%')
```
