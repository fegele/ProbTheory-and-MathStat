```python
import math

def count_combinations(n, k):
    return int(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))

def probability(event_outcomes, total_outcomes):
    return event_outcomes/total_outcomes

def Bernoulli(p, n, k):
    C = count_combinations(n, k)
    q = 1 - p
    P = C*(p**k)*(q**(n-k))
    return P

def Poisson(p, n, m):
    lam = p*n
    e = math.exp(1)
    P = (lam**m)/math.factorial(m)*(e**(-lam))
    return P
```

```python
p = 0.8     # вероятность попадания в мишень за один раз
n = 100     # количество выстрелов
k = 85      # число наступлений события

# p > 0.1, количество выстрелов небольшое - используем формулу Бернулли
P_85 = Bernoulli(p, n, k)
print(f'{P_85=}')
```

```python
p = 0.0004    # вероятность перегорания лампочки в течение первого дня
n = 5000      # количество лампочек

# p < 0.1, количество лампочек большое - используем формулу Пуассона
P_0 = Poisson(p, n, 0)
print(f'{P_0=}')
P_2 = Poisson(p, n, 2)
print(f'{P_2=}')
```

```python
### 4a
n1 = 10     # общее число мячей в первом ящике
w1 = 7      # число белых мячей в первом ящике
k1 = 2      # число мячей, вынутых из первого ящика
n2 = 11     # общее число мячей во втором ящике
w2 = 9      # число белых мячей во втором ящике
k2 = 2      # число мячей, вынутых из второго ящика

# число благоприятствующих исходов (2 белых мяча) - 1-й ящик
w1_comb = count_combinations(w1, k1)

# число благоприятствующих исходов (2 белых мяча) - 2-й ящик 
w2_comb = count_combinations(w2, k2)

# общее число исходов
n1_comb = count_combinations(n1, k1)
n2_comb = count_combinations(n2, k2)

# вероятность: 2 мяча из 1-го ящика И 2 мяча из 2-го ящика
P_4 = probability(w1_comb * w2_comb, n1_comb * n2_comb)
print(f'{P_4=}')

# 2-й способ:
print(7/10 * 6/9 * 9/11 * 8/10)
```

```python
### 4b
# варианты благоприятствующих исходов:

# 2 белых мяча из 1-го ящика, 0 из 2-го:
w_comb = count_combinations(w1, 2) * count_combinations(w2, 0)
other_comb = count_combinations(n1-w1, 0) * count_combinations(n2-w2, 2)
comb_1 = w_comb * other_comb

# 1 белый мяч из 1-го ящика, 1 из 2-го:
w_comb = count_combinations(w1, 1) * count_combinations(w2, 1)
other_comb = count_combinations(n1-w1, 1) * count_combinations(n2-w2, 1)
comb_2 = w_comb * other_comb

# 0 белых мячей из 1-го ящика, 2 из 2-го:
w_comb = count_combinations(w1, 0) * count_combinations(w2, 2)
other_comb = count_combinations(n1-w1, 2) * count_combinations(n2-w2, 0)
comb_3 = w_comb * other_comb

# вероятность: 1-й ИЛИ 2-й ИЛИ 3-й вариант
P_2 = probability(comb_1 + comb_2 + comb_3, n1_comb * n2_comb)
print(f'{P_2=}')

# 2-й способ:
P1 = 7/10*6/9*2/11*1/10
P2 = 7/10*3/9*9/11*2/10 + 7/10*3/9*2/11*9/10 + 3/10*7/9*9/11*2/10 + 3/10*7/9*2/11*9/10
P3 = 3/10*2/9*9/11*8/10
P = P1 + P2 + P3
print(P)
```

```python
### 4c
# вероятность извлечения четырёх НЕ белых мячей:
other_comb_1 = count_combinations(n1 - w1, k1)
other_comb_2 = count_combinations(n2 - w2, k2)
other_prob = probability(other_comb_1 * other_comb_2, n1_comb * n2_comb)
print(f'{other_prob=}')

# 2-й способ:
print(3/10 * 2/9 * 2/11 * 1/10)

# вероятность извлечения четырёх мячей, из которых хотя бы один мяч белый:
P = 1 - other_prob
print(f'{P=}')
```
