```python
salaries = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
n = len(salaries)

# среднее арифметическое:
mean_value = sum(salaries)/n
print(f'{mean_value=}')

# среднее квадратичное отклонение:
RMSD = np.sqrt(sum([(x - mean_value) ** 2 for x in salaries])/n)
print(f'{RMSD=}')

# смещённая оценка дисперсии:
biased_estimation = (sum([(x - mean_value) ** 2 for x in salaries])/n)
print(f'{biased_estimation=}')

# несмещённая оценка дисперсии:
unbiased_estimation = (sum([(x - mean_value) ** 2 for x in salaries])/(n-1))
print(f'{unbiased_estimation=}')
```

```python
def count_combinations(n, k):
    return int(np.math.factorial(n)/(np.math.factorial(k)*np.math.factorial(n-k)))

def probability(event_outcomes, total_outcomes):
    return event_outcomes/total_outcomes
```

```python
n1 = 8      # общее число мячей в первом ящике
w1 = 5      # число белых мячей в первом ящике
k1 = 2      # число мячей, вынутых из первого ящика
n2 = 12     # общее число мячей во втором ящике
w2 = 5      # число белых мячей во втором ящике
k2 = 4      # число мячей, вынутых из второго ящика

# варианты благоприятствующих исходов:

# 2 белых мяча из 1-го ящика, 1 из 2-го:
w_comb = count_combinations(w1, 2) * count_combinations(w2, 1)
other_comb = count_combinations(n1-w1, 0) * count_combinations(n2-w2, 3)
comb_1 = w_comb * other_comb

# 1 белый мяч из 1-го ящика, 2 из 2-го:
w_comb = count_combinations(w1, 1) * count_combinations(w2, 2)
other_comb = count_combinations(n1-w1, 1) * count_combinations(n2-w2, 2)
comb_2 = w_comb * other_comb

# 0 белых мячей из 1-го ящика, 3 из 2-го:
w_comb = count_combinations(w1, 0) * count_combinations(w2, 3)
other_comb = count_combinations(n1-w1, 2) * count_combinations(n2-w2, 1)
comb_3 = w_comb * other_comb

# общее число исходов
n1_comb = count_combinations(n1, k1)
n2_comb = count_combinations(n2, k2)

# вероятность: 1-й ИЛИ 2-й ИЛИ 3-й вариант
P_3 = probability(comb_1 + comb_2 + comb_3, n1_comb * n2_comb)
print(f'{P_3=}')
```

```python
p1 = 0.9      # вероятность попадания для 1-го спортсмена
p2 = 0.8      # вероятность попадания для 2-го спортсмена
p3 = 0.6      # вероятность попадания для 3-го спортсмена
p1_A = 1/3    # вероятность выстрела 1-м спортсменом
p2_A = 1/3    # вероятность выстрела 2-м спортсменом
p3_A = 1/3    # вероятность выстрела 3-м спортсменом

# полная вероятность:
P_A = p1*p1_A + p2*p2_A + p3*p3_A
print(f'{P_A=}')

# вероятность того, что выстрел произведён первым спортсменом:
PA_B1 = (p1*p1_A)/P_A
print(f'{PA_B1=}')

# вероятность того, что выстрел произведён вторым спортсменом:
PA_B2 = (p2*p2_A)/P_A
print(f'{PA_B2=}')

# вероятность того, что выстрел произведён третьим спортсменом:
PA_B3 = (p3*p3_A)/P_A
print(f'{PA_B3=}')
```

```python
p1 = 0.8      # вероятность сдать сессию для студента факультета А
p2 = 0.7      # вероятность сдать сессию для студента факультета B
p3 = 0.9      # вероятность сдать сессию для студента факультета C
p1_A = 1/4    # вероятность того, что студент учится на факультете A
p2_A = 1/4    # вероятность того, что студент учится на факультете B
p3_A = 1/2    # вероятность того, что студент учится на факультете C

# полная вероятность:
P_A = p1*p1_A + p2*p2_A + p3*p3_A
print(f'{P_A=}')

# вероятность того, что сдавший первую сессию студент учится на факультете А:
PA_B1 = (p1*p1_A)/P_A
print(f'{PA_B1=}')

# вероятность того, что сдавший первую сессию студент учится на факультете B:
PA_B2 = (p2*p2_A)/P_A
print(f'{PA_B2=}')

# вероятность того, что сдавший первую сессию студент учится на факультете C:
PA_B3 = (p3*p3_A)/P_A
print(f'{PA_B3=}')
```

```python
p1 = 0.1      # вероятность выйти из строя для 1-й детали
p2 = 0.2      # вероятность выйти из строя для 2-й детали
p3 = 0.25     # вероятность выйти из строя для 3-й детали
q1 = 1 - p1   # вероятность НЕ выйти из строя для 1-й детали
q2 = 1 - p2   # вероятность НЕ выйти из строя для 2-й детали
q3 = 1 - p3   # вероятность НЕ выйти из строя для 3-й детали

# вероятность выхода из строя всех деталей:
Pa = p1*p2*p3
print(f'{Pa=}')

# вероятность выхода из строя только двух деталей:
Pb = p1*p2*q3 + p1*q2*p3 + q1*p2*p3
print(f'{Pb=}')

# вероятность выхода из строя хотя бы одной детали:
Pc = 1 - q1*q2*q3
print(f'{Pc=}')

# вероятность выхода из строя от одной до двух деталей:
Pd = Pc - Pa
print(f'{Pd=}')
```
