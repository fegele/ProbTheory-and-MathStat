```python
std = 16
M = 80
n = 256

# Z-критерий
Z = 1.96      # значение Z для α = 0,05
x1 = M - Z * std/np.sqrt(n)
x2 = M + Z * std/np.sqrt(n)
print(f'[{x1};{x2}]')
```

```python
values = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]

n = len(values)
print(f'{n=}')

M = np.mean(values)
print(f'{M=}')

# t-критерий
t = 2.262      # значение t при n = 10 и α = 0,05

S = np.std(values, ddof = 1)
print(f'{S=}')

x1 = M - t * S/np.sqrt(n)
x2 = M + t * S/np.sqrt(n)
print(f'[{x1};{x2}]')
```

```python
# H0: μ = μ0
# H1: μ > μ0

μ0 = 17
μ = 17.5
n = 100
D = 4
std = np.sqrt(D)

Z = (μ - μ0)/(std/np.sqrt(n))
print(f'{Z=}')
```

```python
# Z табличное = 1.65
# полученное Z > Z табличное, H1 верна
```

```python
# H0: μ = μ0
# H1: μ != μ0

μ0 = 200
values = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
n = 10

μ = np.mean(values)
print(f'{μ=}')

S = np.std(values, ddof = 1)
print(f'{S=}')

t = np.abs((μ - μ0)/(S/np.sqrt(n)))
print(f'{t=}')
```

```python
# при n = 10 и α = 0,01 значение t = 3.25
# полученное |t| < t табличное, H0 верна
```
