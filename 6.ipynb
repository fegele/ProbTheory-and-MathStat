```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

cov_1 = np.mean(zp*ks) - np.mean(zp) * np.mean(ks)
print(f"{cov_1=}")
cov_2 = np.cov(zp, ks, ddof=0)
print(f"{cov_2=}")

std_zp = np.sqrt(sum([(x - np.mean(zp))**2 for x in zp])/len(zp))
std_ks = np.sqrt(sum([(x - np.mean(ks))**2 for x in ks])/len(ks))
r_1 = cov_1 / (std_zp * std_ks)
print(f"{r_1=}")
r_2 = np.cov(zp, ks, ddof=0) / (np.std(zp, ddof=0) * np.std(ks, ddof=0))
print(f"{r_2=}")
r_2 = np.cov(zp, ks, ddof=0) / (np.std(zp, ddof=0) * np.std(ks, ddof=0))
print(f"{r_2=}")
```

```python
iq = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])

n = len(iq)
print(f'{n=}')

M = np.mean(iq)
print(f'{M=}')

# t-критерий
t = 2.262      # значение t при n = 10 и α = 0,05

S = np.std(iq, ddof = 1)
print(f'{S=}')

x1 = M - t * S/np.sqrt(n)
x2 = M + t * S/np.sqrt(n)
print(f'[{x1};{x2}]')
```
