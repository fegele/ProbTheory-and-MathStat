```python
import numpy as np
from scipy import stats
```

```python
# H0: рост в среднем одинаков для представителей рассматриваемых видов спорта
# H1: для представителей каких-то видов спорта средний рост отличается от остальных
```

```python
# значения роста в группе футболистов:
g1 = np.array([173, 175, 180, 178, 177, 185, 183, 182])

# значения роста в группе хоккеистов:
g2 = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])

# значения роста в группе штангистов:
g3 = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])

stats.f_oneway(g1, g2, g3)
```

![8-output1](8-output1.png)

```python
# pvalue = 0.010482206918698694
# pvalue < 0.05, среди представителей разных видов спорта есть статистически значимые различия среднего роста
```

```python
# вычисление без использования stats функции:
k = 3
n1 = len(g1)
n2 = len(g2)
n3 = len(g3)
n = n1 + n2 + n3

# средние значения в каждой группе:
y1 = g1.sum()/n1
y2 = g2.sum()/n2
y3 = g3.sum()/n3
print(f"{y1=}")
print(f"{y2=}")
print(f"{y3=}")

# общее среднее значение:
y = (g1.sum() + g2.sum() + g3.sum()) / n
print(f"{y=}")

# сумма квадратов отклонений наблюдений от общего среднего:
S2 = sum([(x-y)**2 for x in g1]) + sum([(x-y)**2 for x in g2]) + sum([(x-y)**2 for x in g3])
print(f"{S2=}")

# сумма квадратов отклонений средних групповых значений от общего среднего:
S2F = (np.mean(g1)-y)**2*n1 + (np.mean(g2)-y)**2*n2 + (np.mean(g3)-y)**2*n3
print(f"{S2F=}")

# остаточная сумма квадратов отклонений:
S2res = sum([(x-np.mean(g1))**2 for x in g1]) + sum([(x-np.mean(g2))**2 for x in g2]) + sum([(x-np.mean(g3))**2 for x in g3])
print(f"{S2res=}")

# сумма квадратов отклонений наблюдений от общего среднего - проверка:
S2 = S2F + S2res
print(f"{S2=}")

# факторная дисперсия:
std2F = S2F / (k - 1)
print(f"{std2F=}")

# остаточная дисперсия:
std2res = S2res / (n - k)
print(f"{std2res=}")

# наблюдаемый критерий Фишера:
FH = std2F/std2res
print(f"{FH=}")
```
